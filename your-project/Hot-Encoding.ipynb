```python
import pandas as pd
```

```python
dt = pd.read_csv('Shelter_animal_outcome_clean')
```

```python
dt.head()
```

# HOT ENCODING FOR BREED

```python
breed_dummy = pd.get_dummies(dt['Breed'])
```

```python
merged = dt.merge(breed_dummy, left_index=True, right_index=True)
merged.head()
```

# HOT ENCODING FOR Color1

```python
#create dummies for color1
col1_dummy = pd.get_dummies(dt['Color1'])
col1_dummy.head()
```

```python
col1_dummy.columns
```

#  HOT ENCODING FOR Color2

```python
col2_dummy = pd.get_dummies(dt['Color2'])

col2_dummy.head()
```

```python
col2_dummy.columns
```

```python
## merge the two color dummies into one 

merg_duplicate = col1_dummy.merge(col2_dummy, left_index=True, right_index=True)
merg_duplicate.head()
```

```python

rename = [x[:-2] if '_' in x else x for x in list(merg_duplicate.columns)]
```

```python
merg_duplicate.columns = rename
```

```python
merg_duplicate.head()
```

```python
"""
The merge duplicate still has the color dubled but now they have the same name so we can group them by name by summing
the values.

"""
```

```python
## Group de columns with the same name and sum them. so if 0 it keeps 0 else I get 1 - kepping the dummies correct 

merged2 = merg_duplicate.groupby(merg_duplicate.columns, axis=1).sum()
merged2.head()
```

```python
## Confirmation : 
```

```python
dt[dt['Color1'] == 'Agouti']
```

```python
merged2[merged2['Agouti'] == 1]
```

```python
df = merged.merge(merged2, left_index=True, right_index=True)
df.head()
```

# # HOT ENCODING FOR GENERE

```python


gen_dummy = pd.get_dummies(dt['Genere'])
```

```python
## merge the two color dummies into one 

dtf = df.merge(gen_dummy, left_index=True, right_index=True)
dtf.head()
```

# Drop columns used for Hot encoding 


```python
dtf.drop(columns = ['AnimalID','Breed', 'Color1','Color2', 'Genere'], inplace = True)
```

```python
## Change Dog to 1 and cat to 0

dtf['AnimalType'] = dtf['AnimalType'].replace('Dog', 1)
dtf['AnimalType'] = dtf['AnimalType'].replace('Cat', 0)
```
